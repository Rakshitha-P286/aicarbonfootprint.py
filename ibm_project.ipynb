```python
%%writefile train_model.py
import pandas as pd
from sklearn.linear_model import LinearRegression
import pickle

data = pd.read_csv("data/carbon_data.csv")

X = data[["electricity_kwh","travel_km","meat_meals"]]
y = data["carbon_kg"]

model = LinearRegression()
model.fit(X, y)

pickle.dump(model, open("model/carbon_model.pkl","wb"))

print("Model trained and saved")
```

```python
%%writefile predict.py
import pickle
import pandas as pd

# Load trained model
model = pickle.load(open("model/carbon_model.pkl", "rb"))

print("\nğŸŒ AI Carbon Footprint Calculator\n")

# User input
electricity = float(input("Electricity usage per month (kWh): "))
travel = float(input("Travel distance per month (km): "))
meat = int(input("Meat meals per week: "))

# Create input
input_data = pd.DataFrame(
    [[electricity, travel, meat]],
    columns=["electricity_kwh", "travel_km", "meat_meals"]
)

# Predict
carbon = model.predict(input_data)[0]

print("\nğŸŒ± Estimated Carbon Footprint:", round(carbon, 2), "kg COâ‚‚")

# AI Sustainability Advice
print("\nğŸ¤– AI Sustainability Advice:")

if electricity > 150:
    print("ğŸ”Œ Reduce electricity usage: switch off unused devices, use LED bulbs.")

if travel > 100:
    print("ğŸš² Reduce travel emissions: prefer public transport or carpooling.")

if meat > 5:
    print("ğŸ¥— Reduce meat consumption: try plant-based meals twice a week.")

if carbon < 150:
    print("âœ… Excellent! Your lifestyle is eco-friendly.")
elif carbon < 250:
    print("âš ï¸ Moderate impact. Small changes can make a big difference.")
else:
    print("âŒ High impact. Immediate lifestyle changes recommended.")
```

```python
%%writefile app.py
import streamlit as st
import pickle
import pandas as pd

# Load model
model = pickle.load(open("model/carbon_model.pkl", "rb"))

st.set_page_config(page_title="AI Carbon Footprint Calculator", page_icon="ğŸŒ")

st.title("ğŸŒ AI Carbon Footprint Calculator")
st.write("Estimate your carbon footprint and get sustainability advice")

# Inputs
electricity = st.slider("Monthly Electricity Usage (kWh)", 0, 500, 150)
travel = st.slider("Monthly Travel Distance (km)", 0, 500, 100)
meat = st.slider("Meat Meals per Week", 0, 14, 4)

# Predict button
if st.button("Calculate Carbon Footprint"):
    input_data = pd.DataFrame(
        [[electricity, travel, meat]],
        columns=["electricity_kwh", "travel_km", "meat_meals"]
    )

    carbon = model.predict(input_data)[0]

    st.subheader("ğŸŒ± Estimated Carbon Footprint")
    st.success(f"{round(carbon, 2)} kg COâ‚‚")

    st.subheader("ğŸ¤– AI Sustainability Advice")

    if electricity > 150:
        st.warning("ğŸ”Œ Reduce electricity usage by using energy-efficient appliances.")
    if travel > 100:
        st.warning("ğŸš² Prefer public transport or carpooling.")
    if meat > 5:
        st.warning("ğŸ¥— Try reducing meat consumption for lower emissions.")

    if carbon < 150:
        st.success("âœ… Excellent! Your lifestyle is eco-friendly.")
    elif carbon < 250:
        st.info("âš ï¸ Moderate footprint. Small changes help.")
    else:
        st.error("âŒ High footprint. Immediate action recommended.")
```
